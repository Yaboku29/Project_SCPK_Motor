```python
import pandas as pd
import numpy as np
df=pd.read_csv('bike_dataset.csv',usecols=lambda col: col!='links')
df=df.set_index('model_name')
print(df.head(10))
```

# Tabel Alternatif & Kriteria


```python
# Alternatif
data=df[(df['type_of_bike']=='Petrol Bike')]
data=data.drop(columns='type_of_bike')
print(data)
# alternatives=petrolBike['model_name']
# print(alternatives)

# Kriteria
# kriteria=['price','CC','mileage','weight_in_kg','acceleration_speed','top_speed']

# Bobot Kriteria
# Price, CC, Mileage, Weight (kg), Accel, Top Speed
bobot={
    'price':0.3,
    'CC':0.1,
    'mileage':0.2,
    'weight_in_kg':0.08,
    'acceleration_speed':0.14,
    'top_speed':0.18
}
# bobot=[0.3,0.1,0.2,0.08,0.14,0.18]
print(bobot)
```

# Penentuan Atribut (Cost & Benefit)

```python
atribut={
    'price':'cost',
    'CC':'benefit',
    'mileage':'cost',
    'weight_in_kg':'cost',
    'acceleration_speed':'benefit',
    'top_speed':'benefit'
}
print(atribut)
```

# Matriks Keputusan


```python
print(data)
```

# Normalisasi Matriks

```python
normalize_matrix=data.copy()
# print(normalize_matrix)

for col in atribut:# Pakai kolom (m)
    if atribut[col]=='benefit':
        normalize_matrix[col]=data[col]/data[col].max()
    else:
        normalize_matrix[col]=data[col].min()/data[col]
print(normalize_matrix)
```

# Preferensi value

```python
preference_value={}

for index,row in normalize_matrix.iterrows():
    total=0
    for k in normalize_matrix.columns:
        total+=row[k] * bobot[k]
    preference_value[index]=total
df_preference_value=pd.DataFrame(preference_value.items(),
                                 columns=['model_name','preference_value'])
df_preference_value=df_preference_value.set_index('model_name')
# print(preference_value)
print(df_preference_value)
```

# Ranking

```python
hasil_ranking=df_preference_value.sort_values(by='preference_value',ascending=False)
print('Hasil Ranking')
print(hasil_ranking)
```
